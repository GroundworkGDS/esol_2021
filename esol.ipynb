```python
import pandas as pd
```

```python
esol = pd.read_csv('TS029-2021-1-filtered-2023-01-27T08_40_31Z.csv')
esol.head()
```

```python
esol_pivoted = esol.pivot(index="Output Areas Code", columns="Proficiency in English language (6 categories)", values="Observation")
esol_pivoted = esol_pivoted.rename(columns = {"Main language is not English (English or Welsh in Wales): Cannot speak English":"cannot_speak_english", "Main language is not English (English or Welsh in Wales): Cannot speak English well": "cannot_speak_english_well"})
esol_pivoted = esol_pivoted.drop(columns =['Main language is English (English or Welsh in Wales)', 'Does not apply', 'Main language is not English (English or Welsh in Wales): Can speak English very well', 'Main language is not English (English or Welsh in Wales): Can speak English well'])
esol_pivoted.head()
```

```python
esol_pivoted['cannot_speak_or_well'] = esol_pivoted.cannot_speak_english + esol_pivoted.cannot_speak_english_well
esol_pivoted = esol_pivoted.drop(columns = ['cannot_speak_english', 'cannot_speak_english_well'])
esol_pivoted.head()
```

```python
esol_pivoted.to_csv('esol_2021_cannot_speak_or_well.csv')
```

OA to Ward

```python
oa_to_ward = pd.read_csv('OA_(2021)_to_Ward_to_LTLA_to_UTLA_to_Region_to_Country_(May_2022)_Lookup_in_England_and_Wales_v2.csv')
oa_to_ward.head()
```

```python
oa_to_ward = oa_to_ward.drop(columns = ['WD22NMW', 'LTLA22NMW', 'UTLA22CD', 'UTLA22NM','UTLA22NMW','RGN22CD', 'RGN22NM', 'RGN22NMW', 'CTRY22CD', 'CTRY22NM', 'CTRY22NMW', 'NAT22CD', 'NAT22NM', 'NAT22NMW', 'ObjectId'])
oa_to_ward = oa_to_ward.rename(columns = {'OA21CD': 'Output Areas Code'})
oa_to_ward.head()
```

```python
oa_to_ward.to_csv('oa_to_ward.csv')
```

```python
merge_oa_ward_borough = esol_pivoted.merge(oa_to_ward, on = 'Output Areas Code', how = 'left')
merge_oa_ward_borough.head()
```

```python
ward = pd.DataFrame()
ward['Count Ward'] = merge_oa_ward_borough.groupby('WD22CD')['cannot_speak_or_well'].sum()
ward.head()
```

```python
borough = pd.DataFrame()
borough['Count Borough'] = merge_oa_ward_borough.groupby('LTLA22CD')['cannot_speak_or_well'].sum()
borough.head()
```

```python
merge_oa_ward_borough = merge_oa_ward_borough.merge(ward, on = 'WD22CD', how = 'left')
merge_oa_ward_borough.head()
```

```python
merge_oa_ward_borough = merge_oa_ward_borough.merge(borough, on = 'LTLA22CD', how = 'left')
```

```python
merge_oa_ward_borough = merge_oa_ward_borough.rename(columns = {'cannot_speak_or_well':'Count OA', 'WD22NM':'Ward Name', 'LTLA22NM':'Borough Name'})
merge_oa_ward_borough = merge_oa_ward_borough[['Output Areas Code', 'Count OA', 'WD22CD', 'Ward Name', 'Count Ward', 'LTLA22CD', 'Borough Name', 'Count Borough']]
merge_oa_ward_borough.head()
```

```python
merge_oa_ward_borough.to_csv('esol_counts_all.csv')
```
